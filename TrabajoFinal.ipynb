##Transformación dataframes

Se convierten los datos al tipo que sea más conveniente y se homogenizan.
En aquellos dataframes que hay nulos si la proporcion no es muy grande es factible eliminar dichas filas.

###Función transformarFechas

Convierte las columnas incluidas en columnas_fechas a datetime

```python
def transformarFechas(df, columnas_fechas) :

  df[columnas_fechas] = df[columnas_fechas].apply(pd.to_datetime)

  return df
```

###Función transformarString

Convierte las columnas incluidas en columnas_string a string

```python
def transformarString(df, columnas_string) :

  for columna in columnas_string :
    df[columna] = df[columna].astype("string")

  return df
```

###Función convertirMayusculas

Convierte las columnas incluidas en columna a mayúsculas

```python
def convertirMayusculas(df, columna) :

  df[columna] = df[columna].str.upper()

  return df
```

###Función convertirMinusculas

Convierte las columnas incluidas en columna a minúsculas

```python
def convertirMinusculas(df, columna) :

  df[columna] = df[columna].str.lower()

  return df
```

##Análisis dataframes

###Función analisisDataFrames

```python
def analisisDF(df) :

  print("Análisis de variables cuantitativas: \n")
  columnas_numericas = df.select_dtypes(include=["int", "float"]).columns  #Identificación de columnas numéricas

  if columnas_numericas.empty :   #Si no tiene columnas numericas lo indica en un mensaje
    print("No posee")
  else :
    print(df[columnas_numericas].describe())  #Si tiene columnas numéricas hace el análisis

  print("----------------------\n")
  print("Análisis de variables cualitativas: \n")
  columnas_string = df.select_dtypes(include=["string"]).columns #Identificación de columnas string

  if columnas_string.empty :  #Si no tiene columnas string lo indica en un mensaje
    print("No posee")
  else :
    print(df[columnas_string].describe())  #Si tiene columnas string hace el análisis

```
